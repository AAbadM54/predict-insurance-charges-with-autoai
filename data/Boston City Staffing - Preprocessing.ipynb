```python
# Dependencies 
import pandas as pd
```

```python
# File names
src_fn = 'boston-data-12cb3883-56f5-47de-afa5-3b1cf61b257b.csv'
dest_fn = 'boston-data-preprocessed-12cb3883-56f5-47de-afa5-3b1cf61b257b.csv'
```

```python
# Load the data into a dataframe
df = pd.read_csv(src_fn)  
```

```python
# Columns 
df.columns
```

```python
# Simple data
df.describe()
```

```python
# Show Head
df.head()
```

```python
# List any INCIDENT_NUMBER with TEST
df = df[~df.INCIDENT_NUMBER.str.contains("TEST", na=False)]
```

```python
# Check the head
df.head()
```

```python
# Drop columns that aren't needed
df = df.drop(columns=['INCIDENT_NUMBER', 'OFFENSE_DESCRIPTION', 'YEAR', 'MONTH', 'Lat', 'Long', 'Location', 'OFFENSE_CODE_GROUP', 'OCCURRED_ON_DATE', 'UCR_PART'])
```

```python
# Show the head
df.head()
```

```python
# OFFENSE_CODE distinct types
df.OFFENSE_CODE.unique()
```

```python
# DISTRICT distinct types
df.DISTRICT.unique()
```

```python
# REPORTING_AREA distinct types
df.REPORTING_AREA.unique()
```

```python
# SHOOTING distinct types
df.SHOOTING.unique()
```

```python
# DAY_OF_WEEK distinct types
df.DAY_OF_WEEK.unique()
```

```python
# HOUR distinct types
df.HOUR.unique()
```

```python
# STREET distinct types
df.STREET.unique()
```

```python
# Create Target Column
df['HOUR_AND_DAY_OF_WEEK'] = df['HOUR'].astype(str) + '_' +  df['DAY_OF_WEEK'].astype(str)
```

```python
# Show the details
df.head()
```

```python
# output to csv
df.to_csv(dest_fn)
```
